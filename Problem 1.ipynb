```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the system parameters
# G(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
```

```python
# Define the transfer function
G = ctrl.TransferFunction(num, den)
print("Transfer Function G(s):")
print(G)
```

```python
#Determine poles and zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
#Generate the Step Response plot
t, y = ctrl.step_response(G)
plt.figure(figsize=(10, 5))
plt.plot(t, y)
plt.title('Step Response of the Two-Mass System')
plt.xlabel('Time (s)')
plt.ylabel('Output (x2)')
plt.grid(True)
plt.show()
```

```python
#Visualize poles and zeros on the complex plane
plt.figure(figsize=(6, 6))
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map')
plt.show()
```
