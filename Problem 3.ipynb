```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  Define the Transfer Function
# G(s) = (4s^2 + 2s + 6) / (16s^4 + 24s^3 + 52s^2 + 36s)
num = [4, 2, 6]
den = [16, 24, 52, 36, 0]
G = ctrl.TransferFunction(num, den)
print ("Transfer Function G(s) =")
```

```python
#  Determine Poles and Zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
#  Generate the Step Response plot
t, y = ctrl.step_response(G)
plt.figure(figsize=(10, 5))
plt.plot(t, y)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid()
plt.show()
```

```python
#  Visualize the poles and zeros on the complex plane
plt.figure(figsize=(10, 5))
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map')
plt.show()
```
