```python
import control as ctrl
import matplotlib.pyplot as plt

num = [5,5]             #5s + 5
den = [1, 4, 3]         # s^2 + 4s + 3
G = ctrl.TransferFunction(num, den)
print(G)

 
```

```python
poles = ctrl.poles(G)
print("Poles of the transfer function:", poles)
```

```python
#Zeros are the roots of the numerator.
zeros = ctrl.zeros(G)
print("Zeros of the transfer function:", zeros)
```

```python
import matplotlib.pyplot as plt
ctrl.pzmap(G, grid=True, title='Pole-Zero Map')
plt.show()
```
