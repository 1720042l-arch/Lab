```python
import serial
import time
import sys
import serial.tools.list_ports
```

```python
def connect_to_arduino(port, baud_rate=9600):
    try:
        ser = serial.Serial(port, baud_rate, timeout=1)
        time.sleep(2) # Wait for reset
        return ser
    except Exception as e:
        print(f"Connection failed: {e}")
        return None
```

```python
def main():
    ports = serial.tools.list_ports.comports()
    for i, port in enumerate(ports):
        print(f"{i}: {port.device}")
    
    choice = input("Select Port Number: ")
    arduino = connect_to_arduino(ports[int(choice)].device)
    if arduino:
        print("Connected to Arduino!")
        arduino.close()
    else:
        print("Failed to connect.")
```

```python
if arduino:
        while True:
            user_input = input("Enter angle (0-180) or 'q': ")
            if user_input.lower() == 'q': break
            
            try:
                angle = int(user_input)
                if 0 <= angle <= 180:
                    arduino.write(f"{angle}\n".encode('utf-8'))
                    print(arduino.readline().decode().strip())
                else:
                    print("Error: Range 0-180 only.")
            except ValueError:
                print("Error: Numeric input required.")
        arduino.close()

if __name__ == "__main__":
    main()
```
