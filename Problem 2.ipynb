```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the transfer function G(s) = 1 / (s^4 + 2*s^3 + 3*s^2 + 2*s)
num = [1]
den = [1, 2, 3, 2, 0]
G = ctrl.TransferFunction(num, den)
```

```python
# Display the Transfer Function
print("Transfer Function G(s) =")
print(G)
```

```python
# Calculate and print Poles and Zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
# Generate the Step Response plot
t, y = ctrl.step_response(G)
plt.figure(figsize=(10, 5))
plt.plot(t, y)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

```python
# Visualize the pole-zero map
plt.figure(figsize=(6, 6))
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map')
plt.show()
```
